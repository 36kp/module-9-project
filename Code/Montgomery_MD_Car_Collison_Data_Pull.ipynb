```python
# url = 'https://apps.bea.gov/api/data'

# query_url = f'{url}?UserID={BEA_Key}&datasetname=Regional&Year=2016&ResultFormat=JSON'

response = requests.get('https://data.montgomerycountymd.gov/api/views/mmzv-x632/rows.json?accessType=DOWNLOAD')
```

```python
json_data = response.json()

```

```python
# print(json.dumps(json_data, indent=4))
```

```python
dictionary_list = json_data['meta']['view']['columns']
# dictionary_list
```

```python
column_names = [dictionary['name'] for dictionary in dictionary_list][8:-9]
# column_names
```

```python
all_data = []
for list_data in json_data['data']:
    data = list_data[8:-9]
    all_data.append({column:value for column, value in zip(column_names,data)})

all_data
```

```python
pd.DataFrame(all_data)
```
